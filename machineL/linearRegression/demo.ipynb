### 正规方程

```python
X= np.array([[1,1],[2,-1]])
X
```

```python
y= np.array([14,10])
y
```

```python
np.linalg.solve(X,y)
```

```python
A=X.T.dot(X)
```

```python
# 逆矩阵  导数
B= np.linalg.inv(A)
B
```

```python
# dot就是矩阵乘法 行 * 列
C=B.dot(X.T)
C
```

```python
C.dot(y)
```

### 三元一次方程

$\begin{cases} x - y +z = 100\\2x +y -z =80\\3x-2y +6z = 256\\\end{cases}$

```python
X=np.array([[1,-1,1],[2,1,-1],[3,-2,6]])
X
y=np.array([100,80,256])
y
```

```python
np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
```

### 带截距的线性方程

$f(X)=x_1+5x_2+15x_3+3x_4+8x_5+4x_6+17x_7+12x_8+12$

```python
y= y+12
y
```

```python

model = LinearRegression(fit_intercept=True)
## fit训练 指定算法
# X与y之间的规律 拟合出来
model.fit(X,y)

b_=model.intercept_
w_=model.coef_
b_
w_
```

```python
X.dot(w_)+b_
```
